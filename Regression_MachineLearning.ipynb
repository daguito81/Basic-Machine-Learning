## Modeling

We will start with the simplest models. Linear Regression  
We will use regular Linear Regression, Ridge and Lasso regressions which use L2 and L1 regularization. Lasoo will be our feature selector

### Data Prep

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
```

```python
X = df.drop('target', axis=1)
y = df['target']
# This splits into our final test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
X_filtered = df[imp_feats].drop('target', axis=1)
# Final test set
X_train_f, X_test_f, y_train, y_test = train_test_split(X_filtered, y, test_size=0.1, random_state=101)
```

### Linear Regression

For this model we will use both the unfiltered and filtered datasets and see how they compare to each other to see if feature selection helped us

```python
l_model = LinearRegression()
```

```python
cross_val_score(l_model, X_train, y_train, cv=5)
```

As this is a synthetic data set, we get extremely high scores with a simple linear regression 

```python
l_model.fit(X_train, y_train)
l_model.score(X_test, y_test)
```

We can also see that we're not overfitted from testing on our test set and we get a similar score from our cross validation

We will use **Lasso** to check if it picks every feature or only some. This works because Lasso uses L1 regularization which can make non-significant parameters 0 (Ridge regression uses L2 Regularization which penalizes coefficients but can't make them 0)

```python
l1_model = Lasso()
```

```python
cross_val_score(l1_model, X_train, y_train, cv=5)
```

Scores are slightly lower than then normal linear regresion. We tune the hyperparameter a bit and see if we get something better.

```python
alphas = np.linspace(0.00001, 0.0001, 100)
train_score = []
cv_score = []
test_score = []
for alpha in alphas:
    model = Lasso(alpha=alpha)
    model.fit(X_train, y_train)
    cv_score.append(cross_val_score(model, X_train, y_train).mean())
    train_score.append(model.score(X_train, y_train))
    test_score.append(model.score(X_test, y_test))

plt.subplots(figsize=(10,6))
plt.plot(alphas, train_score, color='blue', label='Train')
plt.plot(alphas, test_score, color='red', label='Test')
plt.plot(alphas, cv_score, color='green', label='CV')
plt.title("Train / Test / CV Scores Lasso Regression")
#plt.xscale('log')
plt.legend()
```

We can see that the scores are extremely close to each other

```python
sum(model.coef_ != 0)
```

However the Lasso regression picked almost every feature

### Models

```python
# Data Prep
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
l_model = LinearRegression()
```

```python
cross_val_score(l_model, X_train, y_train, cv=5)
```

We can now see that our model is not as good as the previous one with the synthetic data. Which means it's much closer to a real scenario. 

Sklearn doesn't provide p-values as it tries to avoid doing inferential studies. For that we need the stats module in python

```python
import statsmodels.api as sm
```

```python
X2 = sm.add_constant(X_train)
est = sm.OLS(y_train, X2)
fit_model = est.fit()
```

```python
fit_model.summary()
```

We get a 73 R2 and we can see the p-values of the different features, denoting that if they are less than 0.05 they are significant

We use better models within the linear family to see if we can get a better score.

```python
r_model = Ridge()
la_model = Lasso()
```

```python
cross_val_score(r_model, X_train, y_train, cv=5)
```

```python
cross_val_score(la_model, X_train, y_train, cv=5)
```

We're not getting better results with regularization. So we try optimizing the hyper parameters

#### Ridge

```python
alphas = np.logspace(-5,5, 100)
train_score = []
cv_score = []
test_score = []
for alpha in alphas:
    model = Ridge(alpha=alpha)
    model.fit(X_train, y_train)
    cv_score.append(cross_val_score(model, X_train, y_train).mean())
    train_score.append(model.score(X_train, y_train))
    test_score.append(model.score(X_test, y_test))

plt.subplots(figsize=(10,6))
plt.plot(alphas, train_score, color='blue', label='Train')
plt.plot(alphas, test_score, color='red', label='Test')
plt.plot(alphas, cv_score, color='green', label='CV')
plt.title("Train / Test / CV Scores Ridge Regression")
plt.xlabel('alpha')
plt.ylabel('r2')
plt.xscale('log')
plt.legend()
```

We can see that our score drops dramatically with alphas above 10 


#### Lasso

We try the same with Lasso Regression and see if it makes any difference

```python
alphas = np.logspace(-5,5, 100)
train_score = []
cv_score = []
test_score = []
for alpha in alphas:
    model = Lasso(alpha=alpha)
    model.fit(X_train, y_train)
    cv_score.append(cross_val_score(model, X_train, y_train).mean())
    train_score.append(model.score(X_train, y_train))
    test_score.append(model.score(X_test, y_test))

plt.subplots(figsize=(10,6))
plt.plot(alphas, train_score, color='blue', label='Train')
plt.plot(alphas, test_score, color='red', label='Test')
plt.plot(alphas, cv_score, color='green', label='CV')
plt.title("Train / Test / CV Scores Lasso Regression")
plt.xlabel('alpha')
plt.ylabel('r2')
plt.xscale('log')
plt.legend()
```

Seems even worse than Ridge regression and it doesn't seem like we're getting any benefit from this regularization

We need to start looking at different algorithms to see if there is an improvement

### Random Forest and Boosted Trees

Although tree methods are not the best for regression problems due to their if/else nature, we will try them 

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

We first try with a default model

```python
rf_model = RandomForestRegressor()
gb_model = GradientBoostingRegressor()
```

**Random Forest**

```python
cross_val_score(rf_model, X_train, y_train, cv=5)
```

```python
rf_model.fit(X_train, y_train)
print("Train: ", rf_model.score(X_train, y_train))
print("Test: ", rf_model.score(X_test, y_test))
```

**GradientBoosted**

```python
cross_val_score(gb_model, X_train, y_train, cv=5)
```

```python
gb_model.fit(X_train, y_train)
print("Train: ", gb_model.score(X_train, y_train))
print("Test: ", gb_model.score(X_test, y_test))
```

#### GridSearch

We get much better results with the tree methods. We're still going to optimize some hyper parameters to see if we can make it overfit a bit less. But this time we will use GridSearch and Randomized Search to find the optimal parameters

```python
param_grid = {
    'n_estimators': [100, 200, 500, 700, 1000],
    'max_depth': [5, 10, 15, 20, 25, 30, 50],
    'max_features': [None, 'sqrt', 'log2'],
    'min_samples_leaf': [1, 10, 20],
}
```

```python
cv_grid = GridSearchCV(RandomForestRegressor(), param_grid=param_grid, n_jobs=-1, cv=5, verbose=1)
```

```python
cv_grid.fit(X_train, y_train)
```

```python
cv_grid.best_score_
```

```python
plt.boxplot(cv_grid.cv_results_['mean_test_score'])
```

We can see a difference between the grid search (1500 models) and the default train/test split and how different the scores are. That's why Cross Validation is very important. The original score was simply a lucky train/test split

```python
best_rf = cv_grid.best_estimator_
```

```python
best_rf.score(X_train, y_train)
```

```python
best_rf.score(X_test, y_test)
```

There seems to be a bit of overfitting still

**Gradient Boosted Trees**

```python
param_grid = {
    'n_estimators': [100, 200, 500, 700, 1000],
    'subsample': [1.0, 0.9, 0.7, 0.5],
    'max_depth': [3, 5, 7, 10],
    'min_samples_leaf': [1, 10, 20],
}
```

```python
cv_rand = RandomizedSearchCV(GradientBoostingRegressor(), param_distributions=param_grid, n_iter=100, n_jobs=-1, cv=5, verbose=1)
```

```python
cv_rand.fit(X_train, y_train)
```

```python
cv_rand.best_score_
```

```python
cv_rand.best_params_
```

```python
best_gb = cv_rand.best_estimator_
```

```python
best_gb.score(X_train, y_train)
```

```python
best_gb.score(X_test, y_test)
```

We get  a much better model with gradient boosted trees. So we will use that one for our predictions
